```python
#option of parse_dates
data2 = pd.read_csv("aapl.txt",sep=",",parse_dates=['Date'])
```

##  RESAMPLING 

```python
data2['Close'].resample('M').mean()
```

```python
data2['Close'].resample('M').mean().plot()
```

## DATE RANGE

```python
data_no_dates = pd.read_csv("aapl_no_dates.txt",sep=",")
data_no_dates.head(3)
```

```python
rang1 = pd.date_range(start='6/1/2017',end='6/30/2017',freq='B')
```

```python
rang1
```

```python
data_no_dates.set_index(rang1, inplace=True)
data_no_dates.head()
```

```python
data_no_dates.Close.plot()
```

```python
#slicing
data_no_dates["2017-06-01":"2017-06-10"]
```
